<h2>Working as a Team</h2>

<p>Building a 3‑level game isn't something one person can do alone. One of the most valuable parts of this project was learning how to <strong>work as a real team</strong>—not just splitting tasks, but actually staying in sync, communicating progress, and holding each other accountable.</p>

<p>We used <strong>GitHub Issues</strong> and a <strong>project board</strong> to manage our workflow. Every feature, bug, and task was tracked as an issue, and we organized them into columns (To Do, In Progress, Done) so everyone could see at a glance what needed attention and who was working on what. This made it easy to:</p>

<ul>
<li>Break the project into manageable pieces and assign ownership</li>
<li>Keep track of each other's progress without constant check‑ins</li>
<li>Stay on schedule by seeing what was blocked or falling behind</li>
<li>Document decisions and discussions directly on the issues</li>
</ul>

<p>Using a board changed how we collaborated. Instead of vague "I'll work on it" promises, we had clear, visible commitments. It also helped when someone got stuck—teammates could jump in because they knew exactly where things stood. This experience taught me that <strong>good tooling and transparency</strong> are just as important as writing good code.</p>

<h2>Pirate Heist – Level 1</h2>

<h3>Game Concept</h3>
<p>In this level, the main character <strong>McArchie</strong> sneaks into a guarded area to steal valuable treasure before anyone notices. The goal is to collect four key items:</p>
<ul>
<li>Chest</li>
<li>Gold</li>
<li>Ruby</li>
<li>Treasure Map</li>
</ul>
<p>McArchie explores the area and interacts with objects to collect these items. Once all four are stolen, the player presses <strong>Esc</strong> to progress to Level 2, where McArchie is discovered and forced into a fight. Level 1 focuses on <strong>exploration and interaction</strong>, while Level 2 introduces combat mechanics.</p>

<h3>Objects Used</h3>
<ul>
<li>Player with 8‑way movement</li>
<li>Background environment</li>
<li>Multiple NPCs with dialogue</li>
<li>Collectible objects (Chest, Gold, Ruby, Treasure Map)</li>
<li>Checklist UI system</li>
</ul>

</div>

<div style="background: #1a1a2e; border-radius: 12px; padding: 20px; margin-bottom: 20px; color: white;">

<h3>Key Technical Decisions</h3>

<h4>Multiple NPC Implementation</h4>
<p>One of the main technical improvements in this level was learning how to implement <strong>multiple NPCs</strong> instead of just one. Each NPC was given:</p>
<ul>
<li>A position in the game world</li>
<li>Dialogue that appears when interacted with</li>
<li>Interaction detection when the player presses <strong>E</strong></li>
</ul>
<p>Previously, I had only worked with a single NPC, so creating several required restructuring how NPC objects were created and stored in the game level. This allowed each NPC to function independently while still using the same dialogue system.</p>

<h4>Reusing Existing Systems</h4>
<p>Instead of rebuilding everything from scratch, I reused several systems from previous work:</p>
<ul>
<li>Character movement for McArchie</li>
<li>Dialogue interactions for NPC conversations</li>
<li>Interaction key logic for pressing E</li>
</ul>
<p>Because these systems were already working, I could focus my time on building new features instead of debugging basic mechanics.</p>

</div>

<div style="background: #1a1a2e; border-radius: 12px; padding: 20px; margin-bottom: 20px; color: white;">

<h3>The Checklist System <span style="color: #f5576c;">(Most Difficult Feature)</span></h3>

<p>The checklist system was the hardest feature to implement. The idea was simple: the player sees a list of items to steal, and once collected, each item gets crossed off.</p>

<p><strong>Example checklist at the start:</strong></p>
<ul>
<li>Chest</li>
<li>Gold</li>
<li>Ruby</li>
<li>Treasure Map</li>
</ul>

<p><strong>After collecting an item:</strong></p>
<ul>
<li><s>Chest</s></li>
<li>Gold</li>
<li>Ruby</li>
<li>Treasure Map</li>
</ul>

<p>Even though the concept seemed simple, the coding was much more complicated. To make it work, I had to:</p>
<ul>
<li>Track whether each item has been collected</li>
<li>Update the checklist UI dynamically</li>
<li>Prevent the player from collecting the same item twice</li>
<li>Connect object interaction with checklist updates</li>
</ul>
<p>Overall, the checklist system took the most time out of everything in this level.</p>

</div>

<div style="background: #1a1a2e; border-radius: 12px; padding: 20px; margin-bottom: 20px; color: white;">

<h3>Object Positioning Challenges</h3>
<p>Another surprisingly difficult part was <strong>positioning objects</strong> in the game world. Every object—NPCs, treasure items, and interactable objects—needed exact coordinates. If positioned incorrectly:</p>
<ul>
<li>Items could appear off screen</li>
<li>NPCs could spawn inside walls or other objects</li>
<li>Interaction zones wouldn't align correctly with the sprite</li>
</ul>
<p>This required a lot of trial and error—adjusting coordinates, reloading, and testing again until everything appeared in the correct place. Although it sounds simple, it became one of the more time‑consuming parts of building the level.</p>

<h3>Preparing for Level 2</h3>
<p>After the player collects all four items, the checklist is fully completed, McArchie has stolen the treasure, and the story progresses to <strong>Level 2</strong>—where McArchie is caught and must fight his way out, introducing combat gameplay that builds on the exploration mechanics from Level 1.</p>
